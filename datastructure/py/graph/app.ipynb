```python
import pandas as pd
import numpy as np
```

data format
- row 1 : 节点索引
- row 2 : 节点值
- row 3...end : 节点距离

```python
data = pd.read_csv('./data/q1.csv')
need = np.array(data[:1])[0]  # 需求量
distance = np.array(data[1:].fillna(value=-1))  # 距离 # .fillna(value=-1) np.inf
# for i in range(len(need)):
#     distance[i][i] = 0 # 对角线赋0
```

```python
center = 9 
```

```python
from pyecharts import options as opts
from pyecharts.charts import Graph
```

```python
# from time import sleep

nodes = []
links = []
length = len(need) # 顶点数
center_idx=9
for n in range(length):
    nodes.append(opts.GraphNode(name=f"{n+1}", symbol_size=need[n] / 4, value=need[n], 
    label_opts=opts.LabelOpts(color=('white' if n+1 != center_idx else 'yellow'))))
    for m in range(n+1, length):
        if distance[n][m] > 0:
            links.append(opts.GraphLink(source=f"{n+1}", target=f"{m+1}", value=distance[n][m], 
            linestyle_opts=opts.LineStyleOpts(type_="solid"))) # 虚线 dot
```

```python
print(len(nodes), len(links))
```

```python
graph = (
    Graph()
    .add("", nodes, links, repulsion=1000,
        edge_label=opts.LabelOpts(is_show=True, position="middle", formatter="{c}"),
        label_opts=opts.LabelOpts(is_show=True, formatter="{b}({c})"),
        )
    .set_global_opts(title_opts=opts.TitleOpts(title="graph"))
)
graph.render_notebook()
```

```python
data_graph = []
for n in range(length):
    for m in range(n + 1, length):
        if n != m and distance[n][m] != -1:
            data_graph.append({n+1, m+1, distance[n][m]})
data_graph  # 生成边权矩阵 (x,y->z)
```

```python
def Floyf(_graph,N):
    for k in range(N):
        for i in range(N):
            for j in range(N):
                if i==j:
                    continue
                if _graph[i][k] != -1 and _graph[k][j] != -1:
                    _graph[i][j] = (_graph[i][k] + _graph[k][j]) if _graph[i][j] == -1 else min(_graph[i][j], _graph[i][k] + _graph[k][j])
    return _graph
```
